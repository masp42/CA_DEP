F) Apply dummy encoding to categorical variables (at least one variable use from the data set) and
discuss the benefits of dummy encoding to understand the categorical data.

## Discussion about the benefits of dummy encoding  to understand the categorical data.
### Done by: Marcelo Pugliesi

Dummy encoding make the categorical variable into binarie variable ( with value of zero or one). The objective is to make comparisons of variables easier, providing one way of using categorical predictor variables. Dummy variables allows easy interpretation and calculation of the odds ratios, and increases the stability and significance of the coefficients.

But this is only effective if there are not many categories within the same variable. Dummy encoding is also not effective if there is a dataset with multiple categorical.

In the two cases above, dummy coding ends up creating multiple dummy features in the dataset without adding much information.

H) What is the purpose of dimensionality reduction? Explore the situations where you can gain the
benefit of dimensionality reduction for data analysis.

The purpose of dimensionality reduction is reduce the number of input data variables from a large dataset, capturing the essence of the data without loose accuracy.  

When a dataset is too large and is necessary to extract more relevant data to easily interprete the results, dimensionality reduction is beneficial to bring attention on the most relevant variables. 

Dimensionality reduction reduces the time and space required to storage a dataset, helps to provide a simpler explanation of the dataset and improves the interpretation of machine learning model.

Those benefits listed above can save millions in cost on running a real world large machine learning model.
